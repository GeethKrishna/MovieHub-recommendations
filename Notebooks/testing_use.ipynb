```python
import pickle

# Load DataFrame from the saved file
with open('../Notebooks/pickles/web_200.pkl', 'rb') as file:
    new_df = pickle.load(file)
```

```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
```

```python
def recommed(name):
    Top_movies = []
    cosine_array = []
    index = new_df.loc[new_df['title'] == name].index[0]
    for i in new_df.index:
        A = new_df['genre_vector'][i]
        B = new_df['genre_vector'][index]

        # Check for missing or invalid data (NoneType or empty vectors)
        if A is not None and B is not None and len(A) > 0 and len(B) > 0:
            A = np.array(A)
            B = np.array(B)
            cos = np.dot(A, B) / (norm(A) * norm(B))
            cosine_array.append(cos)
        else:
            cosine_array.append(0.0)  # Handle missing or invalid data with a default value

    # Now cosine_array contains the cosine similarities
    #print(cosine_array)
    sorted_with_indices = sorted(enumerate(cosine_array), key=lambda x: x[1], reverse=True)
    for i in range(1,11):
        Top_movies.append(new_df['title'][sorted_with_indices[i][0]])
    print(Top_movies)
```

```python
recommed('Titanic')
```
